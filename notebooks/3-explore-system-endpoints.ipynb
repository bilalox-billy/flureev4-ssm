# System Endpoints
## Check the health of given endpoint
```json
GET /health
```

Health check endpoint for monitoring

````python
import requests
url = f"http://localhost:8090/health"
response = requests.get(url)
print(response.json())
print(response.status_code)
````

## Detailed server statistics
to get statistics about the given server
```json
GET /stats
```



````python
url = "http://localhost:8090/v1/fluree/stats"

response = requests.get(url)
print(response.json())
print(response.status_code)
````
